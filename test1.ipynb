```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("transaction_1.csv")

# =========================
# 1. INFORMASI UMUM DATA
# =========================
print("Shape Data:", df.shape)
print("\nKolom:")
print(df.columns)

print("\nInfo Data:")
print(df.info())

print("\n5 Baris Pertama:")
display(df.head())

# =========================
# 2. CEK MISSING VALUE
# =========================
missing = df.isnull().sum()

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing Percentage (%)': (missing / len(df)) * 100
})

missing_df = missing_df.sort_values(
    by='Missing Percentage (%)',
    ascending=False
)

display(missing_df)

# =========================
# 3. CEK DUPLIKAT
# =========================
print("Jumlah Duplikat:", df.duplicated().sum())

# =========================
# 4. STATISTIK DESKRIPTIF
# =========================
display(df.describe())

display(df.describe(include='object'))

# =========================
# 5. CEK UNIQUE VALUE
# =========================
for col in df.columns:
    print(f"\n{col}")
    print("Jumlah Unique:", df[col].nunique())

# =========================
# 6. DISTRIBUSI NUMERIK
# =========================
numeric_cols = df.select_dtypes(include=np.number).columns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    df[col].hist(bins=30)
    plt.title(col)
    plt.show()

# =========================
# 7. OUTLIER CHECK
# =========================
for col in numeric_cols:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col].dropna())
    plt.title(col)
    plt.show()

# =========================
# 8. CORRELATION
# =========================
corr = df[numeric_cols].corr()

plt.figure(figsize=(10,8))
plt.imshow(corr, cmap='coolwarm')
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Matrix")
plt.show()
```

```python
import matplotlib.pyplot as plt

promo_count = df['Promo_Category'].value_counts()

plt.figure(figsize=(8,5))
promo_count.plot(kind='bar')
plt.title('Jumlah Transaksi per Promo')
plt.ylabel('Count')
plt.show()

print(promo_count)
```

```python
brand_sales = df.groupby('Brand')['Total_Price'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
brand_sales.plot(kind='bar')
plt.title('Revenue per Brand')
plt.ylabel('Revenue')
plt.show()

display(brand_sales)
```

```python
category_sales = (
    df.groupby('SKU_Category')['Total_Price']
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(10,5))
category_sales.plot(kind='bar')
plt.title('Revenue per Category')
plt.ylabel('Revenue')
plt.show()
```

```python
top_products = (
    df.groupby('SKU')['Qty']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_products.plot(kind='bar')
plt.title('Top 10 Product by Quantity')
plt.show()

display(top_products)
```

```python
top_revenue = (
    df.groupby('SKU')['Total_Price']
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(10,5))
top_revenue.plot(kind='bar')
plt.title('Top 10 Product by Revenue')
plt.show()

display(top_revenue)
```

```python
plt.figure(figsize=(8,5))
df['DiscountPercentage'].hist(bins=20)
plt.title('Distribution of Discount Percentage')
plt.show()
```

```python
promo_effectiveness = (
    df.groupby('Promo_Category')
      .agg({
          'Qty':'mean',
          'Total_Price':'mean',
          'TransactionID':'count'
      })
      .rename(columns={
          'Qty':'Avg_Qty',
          'Total_Price':'Avg_Revenue',
          'TransactionID':'Transaction_Count'
      })
)

display(promo_effectiveness)
```
